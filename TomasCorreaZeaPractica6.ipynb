# Práctica 6: Problemas de valor inicial para EDOs

## Método de Euler

```
import numpy as np
import matplotlib.pyplot as plt

def euler(f, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        wtemp = w[i] + h * f(t[i], w[i])
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return np.exp(t - x)

[w, t] = euler(f, 0, 1, 2, 1)
print(t)
print(w)
```

```
def f(t, x):
    return 1 + (t - x)**2

[w, t] = euler(f, 2, 3, 2, 1)
print(t)
print(w)
```

```
def f(t, x):
    return x - t**2 + 1

[w, t] = euler(f, 0, 2, 10, 0.5)
print(w)
```

```
def f(t, x):
    return t**(-2) * (np.sin(2*t) - 2*t*x)

[w, t] = euler(f, 1, 2, 4, 2)
print(w)
```

## Método de Taylor orden 2

```
def taylor2(f, pf, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        wtemp = w[i] + h*f(t[i], w[i]) + 0.5*h**2*pf(t[i], w[i])
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return x - t**2 + 1

def pf(t, x):
    return x - t**2 - 2*t + 1

[w, t] = taylor2(f, pf, 0, 2, 10, 0.5)
print(t)
print(w)
```

## Método de Taylor orden 4

```
def taylor4(f, df, sd, td, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        wtemp = (w[i]
                 + h*f(t[i], w[i])
                 + 0.5*h**2*df(t[i], w[i])
                 + (1/6)*h**3*sd(t[i], w[i])
                 + (1/24)*h**4*td(t[i], w[i]))
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return x - t**2 + 1

def df(t, x):
    return x - t**2 + 1 - 2*t

def sd(t, x):
    return x - t**2 - 1 - 2*t

def td(t, x):
    return x - t**2 - 1 - 2*t

[w, t] = taylor4(f, df, sd, td, 0, 2, 10, 0.5)
print(w)
```

## Runge-Kutta orden 4

```
def kuttaorden4(f, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        k1 = h * f(t[i], w[i])
        k2 = h * f(t[i] + 0.5*h, w[i] + 0.5*k1)
        k3 = h * f(t[i] + 0.5*h, w[i] + 0.5*k2)
        k4 = h * f(t[i] + h, w[i] + k3)
        wtemp = w[i] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return -x + t * x**0.5

[w, t] = kuttaorden4(f, 2, 3, 4, 2)
print(w)
```

### Método del Punto Medio

```
def punto_medio(f, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        wtemp = w[i] + h * f(t[i] + h/2, w[i] + (h/2)*f(t[i], w[i]))
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return x - t**2 + 1

[w, t] = punto_medio(f, 0, 2, 10, 0.5)
print(t)
print(w)
```

### Método de Euler Modificado

```
def euler_mod(f, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        k1 = h * f(t[i], w[i])
        k2 = h * f(t[i] + h, w[i] + k1)
        wtemp = w[i] + (k1 + k2) / 2
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return x - t**2 + 1

[w, t] = euler_mod(f, 0, 2, 10, 0.5)
print(t)
print(w)
```

### Método de Heun

```
def heun(f, a, b, n, w0):
    h = (b - a) / n
    t = [a]
    w = [w0]
    for i in range(n):
        k1 = h * f(t[i], w[i])
        k2 = h * f(t[i] + h/3, w[i] + k1/3)
        k3 = h * f(t[i] + 2*h/3, w[i] + 2*k2/3)
        wtemp = w[i] + (k1 + 3*k3) / 4
        ttemp = t[i] + h
        w.append(wtemp)
        t.append(ttemp)
    return np.array(w), np.array(t)
```

```
def f(t, x):
    return x - t**2 + 1

[w, t] = heun(f, 0, 2, 10, 0.5)
print(t)
print(w)
```

### Ejercicio 1 — comparación de métodos: y' = te^(3t) - 2y

```
def f(t, x):
    return t * np.exp(3*t) - 2*x

a, b, n, w0 = 0, 1, 10, 0

we, te = euler(f, a, b, n, w0)
wpm, _ = punto_medio(f, a, b, n, w0)
wem, _ = euler_mod(f, a, b, n, w0)
wrk, _ = kuttaorden4(f, a, b, n, w0)

plt.plot(te, we, label='Euler')
plt.plot(te, wpm, label='Punto Medio')
plt.plot(te, wem, label='Euler Mod.')
plt.plot(te, wrk, '--', label='RK4')
plt.legend()
plt.grid()
plt.title("y' = te^(3t) - 2y")
plt.show()
```

### Ejercicio 2 — comparación de métodos: y' = 1 + (t-y)²

```
def f(t, x):
    return 1 + (t - x)**2

a, b, n, w0 = 2, 3, 10, 1

we, te = euler(f, a, b, n, w0)
wpm, _ = punto_medio(f, a, b, n, w0)
wem, _ = euler_mod(f, a, b, n, w0)
wrk, _ = kuttaorden4(f, a, b, n, w0)

plt.plot(te, we, label='Euler')
plt.plot(te, wpm, label='Punto Medio')
plt.plot(te, wem, label='Euler Mod.')
plt.plot(te, wrk, '--', label='RK4')
plt.legend()
plt.grid()
plt.title("y' = 1 + (t-y)²")
plt.show()
```
