## Interpolación de Lagrange

```
%matplotlib inline
import sympy
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 0.6, 0.9])
y = np.array([1, 0.825337, 0.62161])
x_sym = sympy.symbols('x')

def prod(i, j):
    return (x_sym - x[i]) / (x[j] - x[i]) if i != j else 1

points = len(x)
lj = [np.prod([prod(i, j) for i in range(points)]) for j in range(points)]
pol = sum(y * lj)
print('polinomio generado:')
sympy.expand(pol).evalf(n=6)
```

```
x = np.array([-1, 0, 2])
y = np.array([4, 1, -1])
x_sym = sympy.symbols('x')
lj = [np.prod([prod(i, j) for i in range(len(x))]) for j in range(len(x))]
pol = sum(y * lj)
sympy.expand(pol).evalf(n=6)
```

```
x = np.array([0, 3, 5, 8])
y = np.array([0, 225, 383, 623])
x_sym = sympy.symbols('x')
lj = [np.prod([prod(i, j) for i in range(len(x))]) for j in range(len(x))]
pol = sum(y * lj)
print('polinomio generado:')
sympy.expand(pol).evalf(n=6)
```

```
min_value = min(x) - 0.1
max_value = max(x) + 0.1
x_test = np.linspace(min_value, max_value, 100)
y_pol = [pol.subs(x_sym, i) for i in x_test]
plt.plot(x_test, y_pol, 'black')
plt.scatter(x, y, color='r')
plt.legend(['Lagrange', 'Puntos'], loc='best')
plt.show()
```

## Diferencias divididas de Newton

```
from numpy import *

def coeficiente(x, y):
    n = len(x)
    a = []
    for i in range(0, n):
        a.append(y[i])
    for j in range(1, n):
        for i in range(n-1, j-1, -1):
            a[i] = float(a[i] - a[i-1]) / float(x[i] - x[i-j])
    return np.array(a)

def Eval(a, x, r):
    n = len(a) - 1
    temp = a[n]
    for i in range(n - 1, -1, -1):
        temp = temp * (r - x[i]) + a[i]
    return temp
```

```
x = np.array([8.1, 8.3, 8.6, 8.7])
y = np.array([16.94410, 17.56492, 18.50515, 18.82091])
a = coeficiente(x, y)
print(a)
```

### Lagrange como función encapsulada

```
def lagrange(x, y, r):
    n = len(x)
    result = 0
    for j in range(n):
        term = y[j]
        for i in range(n):
            if i != j:
                term *= (r - x[i]) / (x[j] - x[i])
        result += term
    return result
```

### Ejercicio 1 Lagrange — sin(x)

```
x = np.array([0, np.pi/6, np.pi/4, np.pi/3])
y = np.sin(x)
r = np.pi / 5
aprox = lagrange(x, y, r)
print('Aproximacion:', aprox)
print('Valor real:  ', np.sin(r))
print('Error:       ', abs(np.sin(r) - aprox))
```

### Ejercicio 2 Lagrange — e^x

```
x = np.array([0.0, 0.25, 0.5, 0.75])
y = np.exp(x)
r = 0.43
aprox = lagrange(x, y, r)
print('Aproximacion:', aprox)
print('Valor real:  ', np.exp(r))
print('Error:       ', abs(np.exp(r) - aprox))
```

### Ejercicio 1 Diferencias Divididas — sin(x)

```
x = np.array([0, np.pi/6, np.pi/4, np.pi/3])
y = np.sin(x)
a = coeficiente(x, y)
aprox = Eval(a, x, np.pi/5)
print('Aproximacion:', aprox)
print('Valor real:  ', np.sin(np.pi/5))
print('Error:       ', abs(np.sin(np.pi/5) - aprox))
```
