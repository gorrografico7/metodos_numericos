# Práctica 3: Raíces de ecuaciones

## Método de Bisección

```
import numpy as np

def bisec(f, a, b, tol):
    error = 10
    while error > tol:
        p1 = (a + b) / 2
        fa = f(a)
        fp1 = f(p1)
        if fp1 == 0:
            raiz = p1
            break
        elif fa * fp1 < 0:
            b = p1
        else:
            a = p1
        raiz = p1
        error = abs(fp1)
    print(abs(raiz))
```

```
def f(x):
    return x**0.5 - np.cos(x)

bisec(f, 0, 1, 1e-2)
```

```
def f(x):
    return x**4 + 2*x**3 - 2

bisec(f, 0, 1, 1e-3)
```

```
def f(x):
    return x**3 - 25

bisec(f, 2, 3, 1e-3)
```

## Método de Punto Fijo

```
def puntofijo(g, p0, tol):
    error = 10
    while error > tol:
        p = g(p0)
        error = abs(p - p0)
        p0 = p
        print(p0)
```

```
def g(x):
    return (x + 3) / (x + 1)

puntofijo(g, 1, 1e-4)
```

## Método de Newton

```
def newton(f, fp, p0, tol):
    error = 10
    while error > tol:
        p = p0 - f(p0) / fp(p0)
        error = abs(p - p0)
        p0 = p
        print(p0)
```

```
def f(x):
    return x**2 - 6

def df(x):
    return 2*x

newton(f, df, 1, 1e-2)
```

```
def f(x):
    return x**3 - np.cos(x)

def df(x):
    return 3*x**2 + np.sin(x)

newton(f, df, 1, 1e-4)
```

```
import matplotlib.pyplot as plt

p0 = 1
x = np.linspace(-2, 2, 100)
plt.plot(x, f(x))
plt.plot(p0, f(p0), 'or')
plt.grid()
plt.show()
```

### Newton con ciclo for

```
def newton_for(f, fp, p0, tol, n=100):
    for i in range(n):
        p = p0 - f(p0) / fp(p0)
        if abs(p - p0) < tol:
            print(p)
            break
        p0 = p
        print(p0)
```

```
def f(x):
    return x**3 - x - 1

def df(x):
    return 3*x**2 - 1

newton_for(f, df, 1.5, 1e-6)
```

```
def f(x):
    return np.cos(x) - x

def df(x):
    return -np.sin(x) - 1

newton_for(f, df, 0.5, 1e-6)
```

### Método de la Secante

```
def secante(f, p0, p1, tol):
    error = 10
    while error > tol:
        p = p1 - f(p1) * (p1 - p0) / (f(p1) - f(p0))
        error = abs(p - p1)
        p0 = p1
        p1 = p
        print(p)
```

```
def f(x):
    return x**3 - x - 1

secante(f, 1, 2, 1e-6)
```

```
def f(x):
    return x**2 - 6

secante(f, 2, 3, 1e-6)
```
