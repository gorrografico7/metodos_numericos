## Ciclo while

```
i = 1
while i <= 10:
    print(i)
    i = i + 1
```

```
i = 0
while i <= 104:
    print(i)
    i = i + 8
```

```
i = 1
while i < 11:
    print(i, end=' ')
    i = i + 1
```

```
i = 0
while i <= 104:
    if i % 8 == 0:
        print(i)
    i = i + 1
    i = i + 1
```

## Funciones

```
def cuadrado(x):
    y = x * x
    return y

cuadrado(2)
```

```
def raiz(x):
    y = (x + 1) ** (1/3)
    return y

raiz(1)
```

```
raiz(124)
```

```
def gauss(x):
    y = (x * (x + 1) / 2)
    return y

gauss(100)
```

```
def funcion(x):
    for i in range(0, x):
        if i % 5 == 0 and i % 3 == 0:
            print(i, 'es multiplo de 3 y 5')
        elif i % 5 == 0:
            print(i, 'es multiplo de 5')
        elif i % 3 == 0:
            print(i, 'es multiplo de 3')
        else:
            print(i)

funcion(30)
```

```
def mult(x):
    i = 0
    while i <= x:
        if i % 5 == 0 and i % 3 == 0:
            print(i, 'es multiplo de 3 y 5')
        elif i % 5 == 0:
            print(i, 'es multiplo de 5')
        elif i % 3 == 0:
            print(i, 'es multiplo de 3')
        else:
            print(i)
        i = i + 1

mult(30)
```

## Gráficas

```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0, np.pi, 0.01)
plt.plot(x, x**2)
plt.plot(x, np.cos(x))
plt.plot(x, x/2)
y = np.arange(0, np.pi, 0.1)
plt.plot(y, y**2)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Ilustración de gráficas en Python')
plt.axis([0, np.pi, -1.1, 1.1])
plt.text(1, 0.1, 'texto de prueba')
plt.grid(True)
plt.show()
```

```
def f(x):
    return x*np.cos(x) - 2*x**2 + 3*x - 1

x = np.arange(0.2, 0.3, 0.01)
plt.xlabel(r'LD')
plt.ylabel(r'$f(x)$')
plt.title(r'$f(x)=xcosx-2x^2+3x-1$')
plt.plot(x, f(x))
plt.grid(True)
plt.show()
```
