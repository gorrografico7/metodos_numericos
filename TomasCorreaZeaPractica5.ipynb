# Práctica 5: Diferenciación e Integración Numérica

## Diferenciación Numérica

### Diferencia hacia adelante

```
def dadelante(f, x0, h):
    return (f(x0 + h) - f(x0)) / h
```

### Diferencia hacia atrás

```
def datras(f, x0, h):
    return (f(x0) - f(x0 - h)) / h
```

### Diferencia centrada

```
def dcentrada(f, x0, h):
    return (f(x0 + h) - f(x0 - h)) / (2*h)
```

```
import numpy as np

def f(x):
    return np.cos(x)

print(dadelante(f, 0.5, 0.1))
print(datras(f, 0.5, 0.1))
print(dcentrada(f, 0.5, 0.1))
```

### Ejercicio 1 — sin(x)

```
def f(x):
    return np.sin(x)

print('hacia adelante', dadelante(f,0.5,0.1), dadelante(f,0.6,0.1), dadelante(f,0.7,0.1))
print('hacia atras   ', datras(f,0.5,0.1),    datras(f,0.6,0.1),    datras(f,0.7,0.1))
print('centrada      ', dcentrada(f,0.5,0.1), dcentrada(f,0.6,0.1), dcentrada(f,0.7,0.1))
```

## Integración Numérica

### Regla del Trapecio

```
def trapecio(f, x0, x1):
    return ((x1 - x0) / 2) * (f(x0) + f(x1))
```

```
def f(x):
    return x**4

trapecio(f, 0.5, 1)
```

```
def f(x):
    return np.exp(3*x) * np.sin(2*x)

x1 = np.pi / 4
trapecio(f, 0, x1)
```

```
def f(x):
    return x * np.log(x + 1)

trapecio(f, -0.5, 0)
```

### Regla de Simpson

```
def simpson(f, x0, x2):
    h = (x2 - x0) / 2
    x1 = x0 + h
    return (h / 3) * (f(x0) + 4*f(x1) + f(x2))
```

```
def f(x):
    return x**4

simpson(f, 0.5, 1)
```

```
def f(x):
    return np.exp(3*x) * np.sin(2*x)

x1 = np.pi / 4
simpson(f, 0, x1)
```

```
def f(x):
    return x * np.log(x + 1)

simpson(f, -0.5, 0)
```

### Regla compuesta trapezoidal

```
def trapecio_comp(f, a, b, n):
    h = (b - a) / n
    S = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        x = a + i * h
        S = S + f(x)
    return S * h
```

```
def f(x):
    return 1 + np.sin(x - 1)

n = 12
a = 0
b = np.pi
S = trapecio_comp(f, a, b, n)
print('La integral de f es:', S)
```

```
def f(x):
    return x**4

S = trapecio_comp(f, 0.5, 1, 2)
print('La integral de f es:', S)
```

```
def f(x):
    return np.exp(3*x) * np.sin(2*x)

S = trapecio_comp(f, 0, np.pi/4, 8)
print('La integral de f es:', S)
```

```
def f(x):
    return x * np.log(x + 1)

S = trapecio_comp(f, -0.5, 0, 8)
print('La integral de f es:', S)
```

### Regla compuesta de Simpson

```
def simpson_comp(f, a, b, n):
    if n % 2:
        raise ValueError('n debe ser par (se recibio n=%d)' % n)
    h = (b - a) / n
    Si = f(a) + f(b)
    Sp = 0
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            Si = Si + 2*f(x)
        else:
            Sp = Sp + 4*f(x)
    return (Si + Sp) * h / 3
```

```
def f(x):
    return 1 + np.sin(x - 1)

S = simpson_comp(f, 0, np.pi, 12)
print('La integral de f es:', S)
print('El error es:', abs(4.2221972653 - S))
```

```
def f(x):
    return x**4

S = simpson_comp(f, 0.5, 1, 10)
print('La integral de f es:', S)
print('El error es:', abs(0.19375 - S))
```

```
def f(x):
    return np.exp(3*x) * np.sin(2*x)

S = simpson_comp(f, 0, np.pi/4, 10)
print('La integral de f es:', S)
```

```
def f(x):
    return x * np.log(x + 1)

S = simpson_comp(f, -0.5, 0, 10)
print('La integral de f es:', S)
```

## Tarea

### Segunda derivada

```
def d2(f, x0, h):
    return (f(x0 + h) - 2*f(x0) + f(x0 - h)) / h**2
```

### Ejercicio 1 — segunda derivada de sin(x)

```
def f(x):
    return np.sin(x)

x0, h = 0.5, 0.1
aprox = d2(f, x0, h)
real = -np.sin(x0)
print('Aproximacion:', aprox)
print('Valor real:  ', real)
print('Error:       ', abs(real - aprox))
```

### Ejercicio 2 — segunda derivada de e^x

```
def f(x):
    return np.exp(x)

x0, h = 1.0, 0.1
aprox = d2(f, x0, h)
real = np.exp(x0)
print('Aproximacion:', aprox)
print('Valor real:  ', real)
print('Error:       ', abs(real - aprox))
```
